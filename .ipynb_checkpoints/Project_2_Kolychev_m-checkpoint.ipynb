```python
# Настройки подключения к базе данных
db_config = {'user': 'praktikum_student', # имя пользователя
             'pwd': 'Sdf4$2;d-d30pp', # пароль
             'host': 'rc1b-wcoijxj3yxfsf3fs.mdb.yandexcloud.net',
             'port': 6432, # порт подключения
             'db': 'data-analyst-afisha' # название базы данных
             }

# Формируем строку подключения
connection_string = 'postgresql://{}:{}@{}:{}/{}'.format(
    db_config['user'],
    db_config['pwd'],
    db_config['host'],
    db_config['port'],
    db_config['db'],
)

# Создаем соединение для подключения к базе данных
engine = create_engine(connection_string)

# SQL-запрос для извлечения данных
query = '''
SELECT
	user_id,
	device_type_canonical,
	order_id,
	created_dt_msk AS order_dt,
	created_ts_msk AS order_ts,
	currency_code,
	revenue,
	tickets_count,
	created_dt_msk :: date - DATE(lag(created_dt_msk) OVER(
    PARTITION BY user_id ORDER BY created_dt_msk
  )) AS days_since_prev,
	event_id,
	event_name_code AS event_name,
	event_type_main,
	service_name,
	region_name,
	city_name
FROM
	afisha.purchases p
LEFT JOIN afisha.events e
		USING(event_id)
LEFT JOIN afisha.city c
		USING(city_id)
LEFT JOIN afisha.regions r
		USING(region_id)
WHERE
	device_type_canonical IN ('mobile', 'desktop')
	AND event_type_main != 'фильм'
ORDER BY
	user_id
'''

# Выполняем SQL-запрос и загружаем данные в DataFrame
df = pd.read_sql_query(query, engine)
```

```python
# Функция для подсчета пропущенных значений в датафрейме
def na_count_func(df):
    # Количество пропущенных значений в абсолютных значениях
    count_na = df.isna().sum()
    count_na.name = 'count_na'

    #Количество строк в датафрейме
    total_rows = df.shape[0]

    # Количество пропущенных значений в процентах
    percent_na = (count_na/total_rows * 100).round(2)
    percent_na.name = 'percent_na'

    # Выводим таблицу с количеством и процентом пропущенных значений
    return pd.concat([count_na, percent_na], axis=1)

# Проверяем наличие пропущенных значений в датафрейме
na_count_func(df)
```

Далее понизим разрядность в числовых и временн**ы**х типах данных:

```python
# Определяем столбцы с числовыми и временными типами данных
int_cols = df.select_dtypes(include=['int']).columns
float_cols = df.select_dtypes(include=['float']).columns
datetime_cols = df.select_dtypes(include=['datetime']).columns

# Создаем словарь для хранения информации об изменении типов данных
dict_types = {'column_name': [], 'before': [], 'after': []}

# Понижаем разрядность в столбцах с типом int
for col in int_cols:
    # записываем имя столбца в словарь
    dict_types['column_name'].append(col)
    # записываем тип данных до изменения
    dict_types['before'].append(df[col].dtype)
    
    # понижаем разрядность
    df[col] = pd.to_numeric(df[col], downcast='integer')
    
    # записываем тип данных после изменения
    dict_types['after'].append(df[col].dtype)

# Понижаем разрядность в столбцах с типом float
for col in float_cols:
    dict_types['column_name'].append(col)
    dict_types['before'].append(df[col].dtype)

    df[col] = pd.to_numeric(df[col], downcast='float')

    dict_types['after'].append(df[col].dtype)

# Понижаем разрядность в столбцах с типом datetime
dict_types['column_name'].extend(datetime_cols)
dict_types['before'].extend([df[col].dtype for col in datetime_cols])

df['order_dt'] = df['order_dt'].values.astype('datetime64[D]')
df['order_ts'] = df['order_ts'].values.astype('datetime64[s]')

dict_types['after'].extend([df[col].dtype for col in datetime_cols])

# Выводим информацию об изменении типов данных
pd.DataFrame(dict_types)
```

В большинстве полей с временн**ы**ми и числовыми типами данных мы смогли понизить разраядность, неизменной она осталась только в колонках, хранящих данные о выручке.

Проверим столбцы с номинальными данными: `event_type_main`, `service_name`, `region_name`, `city_name`.

```python
# Функция для вывода информации о столбцах с номинальными данными
def obj_col_info(series):
    
    print(f"Информация о столбце '{series.name}':")

    # Количество уникальных значений
    cnt_unique = series.nunique()

    print(f"- Количество уникальных значений: {cnt_unique}")

    # Распределение значений в абсолютных и относительных величинах
    val_cnt = series.value_counts()
    val_cnt_ratio = (series.value_counts(normalize=True) * 100).round(2)

    # Объединяем две cерии в один датафрейм
    col_info = pd.concat([val_cnt, val_cnt_ratio], axis=1)
    col_info.columns = ['count', 'percent']

    # Выводим распределение значений
    if cnt_unique <= 20:
        print("- Распределение значений:")
        display(col_info)
    else:
        print("- 10 самых часто встречающихся значений:")
        display(col_info.head(10))
        
        print("- 10 самых редко встречающихся значений:")
        display(col_info.tail(10))

# Проверяем столбец event_type_main
obj_col_info(df['event_type_main'])
```

Далее проверим статистические показатели и распределение значений в столбцах `revenue_rub` и `tickets_count`. Начнем с первого:

```python
# Функция для вывода статистических показателей и распределения значений
def stats_info(df,
                 column,
                 bins=25,
                 kde=False,
                 title='',
                 xlabel='Величина',
                 ylabel='Частота'):
    
    # Записываем статистики в переменную и выводим ее
    info = df[column].describe().round(2)
    print(f'Описательные статистики столбца {column}:')
    print(info)

    # Строим гистограмму распределения значений
    plt.figure(figsize=(10, 6))
    sns.histplot(data=df,
                 x=column,
                 bins=bins,
                 color='skyblue',
                 edgecolor='k',
                 kde=kde)

    # Добавляем среднее, медиану и моду на гистограмму
    plt.axvline(info['mean'],
                color='r',
                linestyle='--',
                linewidth=1.5,
                label=f"Среднее значение ({round(info['mean'], 2)})")
    
    plt.axvline(info['50%'],
                color='y',
                linestyle='--',
                linewidth=1.5,
                label=f"Медиана ({round(info['50%'], 2)})")
    
    plt.axvline(df[column].mode(dropna=True)[0],
                color='g',
                linestyle='--',
                linewidth=1.5,
                label=f"Мода ({round(df[column].mode(dropna=True)[0], 2)})")
    
    # Настраиваем гистограмму
    plt.title(f"Распределение значений в столбце {column}" if title == ''
              else title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.grid(alpha=0.5)
    plt.legend()
    plt.show()

    # Строим ящик с усами для выявления выбросов
    plt.figure(figsize=(10, 2))
    sns.boxplot(data=df, x=column)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.grid(alpha=0.5)
    plt.show()

stats_info(df,
             'revenue_rub',
             bins=60,
             xlabel='Выручка, руб.')
```

- Распределение значений выручки ассиметричное, скошено вправо.
- Минимальное значение равно -90.76 руб., максимальное 81174.54 руб., а стандартное отклонение 875.5 руб..
- Среднее значение равно 555.57 руб. а медиана 351.14 руб.
- Значение выручки, которое встречается чаще всего: 0 рублей.
- В данных присутствует большой разброс значений и существенное количество выбросов.

Для того, чтобы выбросы (очень большие суммы, превышающие 99 процентиль) и отрицательные значения не искажали результаты анализа, удалим строки с такими значеми.

Для начала проверим количество строк к удалению:

```python
# Создаем условие для фильтра для удаления экстремальных и
# отрицательных значений выручки
filter_condition = (
    df['revenue_rub'] < 0) |\
        (df['revenue_rub'] > df['revenue_rub'].quantile(0.99))

# Записываем общее количество строк в исходном датафрейме дo удаления
total_rows = df.shape[0]


print(f'Количество строк для удаления: {filter_condition.sum()}\
 ({round(filter_condition.sum() / total_rows * 100, 2)}%)')
```

Количество строк к удалению: 3206 или 1.1%.

Сначала удалим строки, потом заново проверим статистические показатели.

```python
# Удаляем строки с экстремальными и отрицательными значениями выручки
df = df[~filter_condition].copy()

# Проверяем статистические показатели и распределение значений
# после удаления выбросов
stats_info(df,
           'revenue_rub',
             bins=25,
             xlabel='Выручка, руб.',
             kde=True)
```

**Задача 3.1.** Постройте профиль пользователя — для каждого пользователя найдите:

- дату первого и последнего заказа;
- устройство, с которого был сделан первый заказ;
- регион, в котором был сделан первый заказ;
- билетного партнёра, к которому обращались при первом заказе;
- жанр первого посещённого мероприятия (используйте поле `event_type_main`);
- общее количество заказов;
- средняя выручка с одного заказа в рублях;
- среднее количество билетов в заказе;
- среднее время между заказами.

После этого добавьте два бинарных признака:

- `is_two` — совершил ли пользователь 2 и более заказа;
- `is_five` — совершил ли пользователь 5 и более заказов.

**Рекомендация:** перед тем как строить профиль, отсортируйте данные по времени совершения заказа.

---


```python
# Группируем данные по пользователям и вычисляем необходимые показатели
users_info = df.groupby('user_id').agg(
    first_order_dt=('order_dt', 'min'),
    last_order_dt=('order_dt', 'max'),
    first_device_type=('device_type_canonical', 'first'),
    first_region_name=('region_name', 'first'),
    first_service_name=('service_name', 'first'),
    first_event_type=('event_type_main', 'first'),
    total_orders=('order_id', 'count'),
    avg_revenue=('revenue_rub', 'mean'),
    avg_tickets_count=('tickets_count', 'mean'),
    avg_days_since_prev=('days_since_prev', 'mean')
)

# Выводим информацию о полученном датафрейме
display(users_info.head(5))
users_info.info()
```

В полученном датафрейме 21831 строк (как и уникальных пользователей) и 10 колонок. Пропущенные значения содержатся только в поле со средним количеством дней между заказами.

Можно округлить значения в полях `avg_revenue` и `avg_tickets_count` и понизить в них, а также в поле `total_orders`, разрядность.

```python
# Округляем значения в полях avg_revenue и avg_tickets_count
users_info['avg_revenue'] = users_info['avg_revenue'].round(2)
users_info['avg_tickets_count'] = users_info['avg_tickets_count'].round(3)

# Понижаем разрядность в полях avg_revenue и avg_tickets_count
for col in ['avg_revenue', 'avg_tickets_count']:
    users_info[col] = pd.to_numeric(users_info[col], downcast='float')

# Понижаем разрядность в поле total_orders
users_info['total_orders'] = pd.to_numeric(users_info['total_orders'], downcast='integer')
users_info['avg_revenue'] = users_info['avg_revenue'].round(2)
# Выводим информацию о датафрейме после всех преобразований
display(users_info.head(3))
users_info.info()
```

Разрядность успешно понижена.

Добавляем бинарные признаки количества заказов:

```python
# Добавляем бинарные признаки количества заказов
users_info['is_two'] = users_info['total_orders'] >= 2
users_info['is_five'] = users_info['total_orders'] >= 5

# Выводим долю пользователей с двумя и более заказами и пятью и более заказами
users_info[['is_two', 'is_five']].mean().round(4)
```

Более 61% пользователей совершили 2 и более заказов. Пять и более заказов совершили почти 29% пользователей.

При интерпретации результатов учитывайте размер сегментов: если в сегменте мало пользователей (например, десятки), то доли могут быть нестабильными и недостоверными, то есть показывать широкую вариацию значений.

```python
# Функция для вычисления и визуализации доли пользователей,
# совершивших два и более заказов, по сегментам заданной колонки
def return_ratio(df,
                 col,
                 title='',
                 xlabel='Доля возвратившихся клиентов',
                 ylabel='Сегмент',
                 figsize=(8,5)):
    
    # Группируем данные по заданной колонке и вычисляем
    # количество пользователей и долю совершивших два и более заказов
    grouped_df = df.groupby(col).agg(
        count = ('is_two', 'count'),
        ratio = ('is_two', 'mean')
    )

    # Сортируем полученный датафрейм по убыванию количества пользователей
    sorted_df = grouped_df.sort_values('count', ascending=False).head(10)
    
    # Выводим информацию о количестве пользователей и доле
    # совершивших два и более заказов по сегментам заданной колонки
    print('Количество пользователей и доля совершивших два и более заказов')
    print(f'по сегментам колонки {col}:')
    display(sorted_df)

    # Строим горизонтальную столбчатую диаграмму
    plt.figure(figsize=figsize)
    sorted_df['ratio'].sort_values().plot.barh()
    plt.grid(axis='x', alpha=.6)
    plt.title('Доля покупателей, совершивших два и более заказов\n' \
    f'по сегментам колонки {col}' if title=='' else title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.xlim(left=.9 * sorted_df['ratio'].min())

    # Добавляем среднее по выборке на диаграмму
    plt.axvline(df['is_two'].mean(),
                color='r',
                linestyle='--',
                linewidth=1.5,
                label=f"Среднее по выборке \
({round(df['is_two'].mean(), 4)})")
    
    plt.legend()
    plt.show()

# Вычисляем и визуализируем долю пользователей,
# совершивших два и более заказов, по типу первого мероприятия
return_ratio(users_info, 'first_event_type', ylabel='Тип первого мероприятия')
```

**Задача 4.2.1.** Проследите связь между средней выручкой сервиса с заказа и повторными заказами.

- Постройте сравнительные гистограммы распределения средней выручки с билета (`avg_revenue_rub`):
    - для пользователей, совершивших один заказ;
    - для вернувшихся пользователей, совершивших 2 и более заказа.
- Ответьте на вопросы:
    - В каких диапазонах средней выручки концентрируются пользователи из каждой группы?
    - Есть ли различия между группами?

Текст на сером фоне:
    
**Рекомендация:**

1. Используйте одинаковые интервалы (`bins`) и прозрачность (`alpha`), чтобы визуально сопоставить распределения.
2. Задайте параметру `density` значение `True`, чтобы сравнивать форму распределений, даже если число пользователей в группах отличается.

---


```python
# Функция для сравнения плотностей распределения двух серий
def compare_densities(series_1,
                      series_2,
                      bins_width=50,
                      label_1='',
                      label_2='',
                      title='',
                      xlabel='Величина',
                      ylabel='Плотность',
                      figsize=(10,6)):
    
    # Определяем границы и количество бинов для гистограмм
    bins = range(int(min(series_1.min(), series_2.min())),
                 int(max(series_1.max(), series_2.max())) + bins_width,
                 bins_width)
    
    # Строим гистограммы плотностей распределения двух серий
    plt.figure(figsize=figsize)
    series_1.plot.hist(density=True,
                          bins=bins,
                          alpha=0.6,
                          edgecolor='k',
                          label=label_1 if label_1 != ''
                            else series_1.name+' (1)')
    plt.axvline(series_1.mean(), # добавляем среднее первой серии
                color='b',
                linestyle='--',
                linewidth=1.5,
                label=f"Среднее ({label_1 if label_1 != '' else 1}): \
{round(series_1.mean())}")
    
    series_2.plot.hist(density=True,
                          bins=bins,
                          alpha=0.6,
                          edgecolor='k',
                          label=label_2 if label_2 != ''
                            else series_2.name+' (2)')
    plt.axvline(series_2.mean(), # добавляем среднее второй серии
                color='r',
                linestyle=':',
                linewidth=1.5,
                label=f"Среднее ({label_2 if label_2 != '' else 2}): \
{round(series_2.mean())}")
    
    plt.title(title if title != '' else
    f'Сравнение плотностей распределения')
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.grid(alpha=0.5)
    plt.legend()
    plt.show()

    # Строим ящики с усами для выявления выбросов
    plt.figure(figsize=(figsize[0], 2))
    plt.boxplot([series_2, series_1],
                vert=False,
                labels=[label_2 if label_2 != '' else series_2.name,
                        label_1 if label_1 != '' else series_1.name])
    plt.title(title if title != '' else
              f'Сравнение распределений {series_1.name} и {series_2.name}')
    plt.xlabel(xlabel)
    plt.grid(alpha=.5, axis='x')
    plt.show()

# Сравниваем плотности распределения средней выручки на пользователя
# в разрезе количества заказов
compare_densities(users_info['avg_revenue'][users_info['is_two']],
                  users_info['avg_revenue'][~users_info['is_two']],
                  bins_width=60,
                  label_1='2 и более заказов',
                  label_2='1 заказ',
                  title='Сравнение распределений средней выручки \
на пользователя\nв разрезе количества заказов',
                  xlabel='Средняя выручка на пользователя, руб.')
```

Распределение средней выручки пользователей с 1 заказом:
- Пик распределения наблюдается в районе от 0 до 100 рублей.
- Для четверти всех клиентов значение средней выручки лежит в диапазоне от 0 до 140 рублей.
- У каждого второго клиента значение средней выручки не превышает 400 рублей.

Распределение средней выручки пользователей с 2 и более заказов:
- Большинство значений сосредоточено в диапазоне от 260 до 750 рублей.
- Самые часто встречающиеся значения выручки находятся в районе 500 рублей.

Несмотря на то, что среднее значение выручки у обеих групп практически одинаковое, рапределения имеют существенные различия:
- У пользователей с 2 и более заказов пик распределения находится в районе 500 рублей, тогда как у пользователей с 1 заказом пик сильно смещен в сторону нуля.
- Оба распределения имеют хвост справа, но средняя выручка со значением, превышающим 1000 рублей, среди пользователей с 1 заказом встречается чаще, чем среди тех, кто сделал 2 или более заказов.

**Задача 4.4.1:** Проведите корреляционный анализ:

- Проанализируйте полученные результаты. Если полученные значения будут близки к нулю, проверьте разброс данных в `total_orders`. Такое возможно, когда в данных преобладает одно значение: в таком случае корреляционный анализ может показать отсутствие связей. Чтобы этого избежать, выделите сегменты пользователей по полю `total_orders`, а затем повторите корреляционный анализ. Выделите такие сегменты:
    - 1 заказ;
    - от 2 до 4 заказов;
    - от 5 и выше.
- Визуализируйте результат корреляции с помощью тепловой карты.
- Ответьте на вопрос: какие признаки наиболее связаны с количеством заказов?

---

```python
# Создаем матрицу корреляции phik
corr_matrix = users_info.phik_matrix()

# Оставляем столбец с коэффициентами корреляции только для поля total_orders
corr_matrix = corr_matrix.loc[corr_matrix.index != 'total_orders']\
                                                            [['total_orders']]
# Строим тепловую карту коэффициентов корреляции
plt.figure(figsize=(1.5,6))
sns.heatmap(corr_matrix.sort_values('total_orders',ascending=False),
            annot=True,
            fmt='.3f',
            cmap='coolwarm',
            linewidths=0.5,
            cbar=False)

plt.title('Тепловая карта коэффициента phi_k \n для данных total_orders')
plt.xlabel('Количество заказов')
plt.show()
```

- __`is_five` (0.64)__  
Высокая связь — ожидаема. Признак мало что даёт для объяснения поведения, так как он напрямую зависит от total_orders.

- __`first_order_dt` (0.43) и `last_order_dt` (0.42)__  
Клиенты, пришедшие давно и/или остающиеся активными до последнего времени, чаще совершают больше заказов.

__Умеренная связь:__

- __`avg_tickets_count` (0.34)__   
Чтобы понять направление корреляции построим диаграмму рассеяния:

```python
plt.figure(figsize=(10,4))
plt.scatter(users_info['total_orders'], users_info['avg_tickets_count'],
            alpha=.2)
plt.title('Взаимосвязь между количеством заказов и средним количеством\n' \
'билетов в заказе пользователей')
plt.xlabel('Количество заказов')
plt.ylabel('Количество билетов')
plt.grid(alpha=.7)
plt.show()
```
